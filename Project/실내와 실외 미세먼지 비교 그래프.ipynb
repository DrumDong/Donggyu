```python
import pandas as pd
import numpy as np

mean_number= pd.read_csv('C:/Users/ehdrb/Desktop/Spyder/mean_number.csv',encoding='cp949')
inside = pd.read_csv('C:/Users/ehdrb/Desktop/Spyder/1201_0331.csv',encoding='cp949')
Jonglo = pd.read_csv('C:/Users/ehdrb/Desktop/공모전 2/환경기상/일별 최종/종로구/종로5,6가동.csv',encoding='cp949')
```

## 전처리

```python
#측정일의 '-'없애기
inside['측정일']=inside['측정일'].str.replace('-','')
#colnames 바꾸기
inside = inside.rename(columns={'측정일':'days'})
inside = inside.rename(columns={'미세먼지':'실내미세먼지'})
# Jonglo days 타입 바꾸기
Jonglo['days']= Jonglo['days'].astype(str)

#merge시키기
project = pd.merge(inside,Jonglo,how='left',on=['days'])

#필요한 데이터만 뽑기
pj = project[['days','실내미세먼지','pm10_종로5,6가동']]
```

## plot

```python
import matplotlib as plt
pj.plot(label='')
plt.title("미세먼지와 실내미세먼지")
plt
```
